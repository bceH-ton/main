```python
import face_recognition
import cv2
import numpy as np
from PIL import ImageGrab
import time
from openpyxl import Workbook
import os
import win32gui
```

```python
#atList = open('Attendance_list.csv', 'w')

starttime = time.strftime('%Y:%m:%d:%H:%M', (time.localtime(time.time())))

atlist = []

atexcel = Workbook()
atexceln = atexcel.create_sheet('출석로그')

images = [
    "./pic/mingu.jpeg",
    "./pic/minguk.jpeg",
    "./pic/sihyun.jpg",
    "./pic/minjun.jpeg",
    "./pic/sihun.jpeg"
]
#mingu_image = face_recognition.load_image_file("mingu.jpeg")
#mingu_face_encoding = face_recognition.face_encodings(mingu_image)[0]

# Create arrays of known face encodings and their names
known_face_encodings = []

for image in images:
    image = face_recognition.load_image_file(image)
    encode = (face_recognition.face_encodings(image))[0]
    known_face_encodings.append(encode)

known_face_names = [
    "Mingu",
    "Minguk",
    "Sihyun",
    "Minjun",
    "Sihun",
]

# Initialize some variables
face_locations = []
face_encodings = []
face_names = []
process_this_frame = True

n_students = ["Mingu", "Minguk", "Sihyun", "Minjun", "Sihun"]
students = set()

t = {}

while True:
    hwnd = win32gui.FindWindow(None, '카메라')#Zoom 회의
    left, top, right, bot = win32gui.GetWindowRect(hwnd)
    print(left, top, right, bot)
    img = ImageGrab.grab()
    save = "screen.png"
    img.save(save)

    frame = cv2.imread("screen.png")
    frame = frame[0:bot, 0:right]

    # Resize frame of video to 1/4 size for faster face recognition processing
    small_frame = cv2.resize(frame, (0, 0), fx=0.5, fy=0.5)

    # Convert the image from BGR color (which OpenCV uses) to RGB color (which face_recognition uses)
    rgb_small_frame = small_frame[:, :, ::-1]

    # Only process every other frame of video to save time
    if process_this_frame:
        # Find all the faces and face encodings in the current frame of video
        face_locations = face_recognition.face_locations(rgb_small_frame)
        face_encodings = face_recognition.face_encodings(rgb_small_frame, face_locations)

        face_names = []

        for face_encoding in face_encodings:
            # See if the face is a match for the known face(s)
            matches = face_recognition.compare_faces(known_face_encodings, face_encoding)
            name = "Unknown"
            
            face_distances = face_recognition.face_distance(known_face_encodings, face_encoding)
            best_match_index = np.argmin(face_distances)
            if matches[best_match_index]:
                name = known_face_names[best_match_index]



            if not( name in list(students)):
                t[name] = time.strftime('%H:%M:%S', (time.localtime(time.time())))
            students.add(name)
            face_names.append(name)

    process_this_frame = not process_this_frame

    font = cv2.FONT_HERSHEY_DUPLEX
    tmp = 60
    cv2.putText(frame, "[Not attendant]", (0, 30), font, 1.0, (0, 0, 255), 1)
    for name in list(set(n_students) - students):
        cv2.putText(frame, name, (0, tmp), font, 1.0, (0, 0, 255), 1)
        tmp += 30

    tmp = 60
    cv2.putText(frame, "[attendant]", (600, 30), font, 1.0, (0, 255, 0), 1)
    for name in list(students):
        cv2.putText(frame, name, (600, tmp), font, 1.0, (0, 255, 0), 1)
        tmp += 30

    # Display the results
    for (top, right, bottom, left), name in zip(face_locations, face_names):
        # Scale back up face locations since the frame we detected in was scaled to 1/4 size
        top *= 2
        right *= 2
        bottom *= 2
        left *= 2

        # Draw a box around the face
        cv2.rectangle(frame, (left, top), (right, bottom), (0, 0, 255), 2)

        # Draw a label with a name below the face
        cv2.rectangle(frame, (left, bottom - 35), (right, bottom), (0, 0, 255), cv2.FILLED)
        font = cv2.FONT_HERSHEY_DUPLEX
        cv2.putText(frame, name, (left + 6, bottom - 6), font, 1.0, (255, 255, 255), 1)

    # Display the resulting image
    cv2.imshow('Video', frame)

    # Hit 'q' on the keyboard to quit!
    if cv2.waitKey(1)&0xFF == 27:
        break

atexceln = atexcel.active

m = 1
for i in known_face_names:
    atexceln.cell(m, 1, i)
    if i in t.keys():
        atexceln.cell(m, 2, t[i])
    else:
        atexceln.cell(m, 2, "미출석")
    m += 1

p = str("./Attendance/attendance.xlsx")

atexcel.save(p)


cv2.destroyAllWindows()
```
